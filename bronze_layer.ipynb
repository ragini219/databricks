**Dynamic Data Load**

```python
dbutils.widgets.text("file_name", "")
```

```python
file_name = dbutils.widgets.get("file_name")
```

```python
df = spark.readStream.format("cloudFiles")\
            .option("cloudFiles.format", "parquet")\
            .option("cloudFiles.schemaLocation", f"abfss://bronze@azuresadatalake.dfs.core.windows.net/checkpoint_{file_name}")\
            .load(f"abfss://source@azuresadatalake.dfs.core.windows.net/{file_name}")

df.display()
```

```python
df.writeStream.format("parquet")\
    .outputMode("append")\
        .option("checkpointLocation", f"abfss://bronze@azuresadatalake.dfs.core.windows.net/checkpoint_{file_name}")\
        .option("path",f"abfss://bronze@azuresadatalake.dfs.core.windows.net/{file_name}")\
        .trigger(once=True)\
        .start()
```

```python
df = spark.read.format("parquet")\
    .load(f"abfss://bronze@azuresadatalake.dfs.core.windows.net/{file_name}")

display(df)
```
