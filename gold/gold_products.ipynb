# Delta Live Tables (now Lakeflow Declarative Pipelines)

Delta Live Tables, now known as **Lakeflow Declarative Pipelines**, is a Databricks framework for building reliable, declarative data pipelines. It manages the flow of data between Delta tables, automating orchestration, cluster management, monitoring, and data quality.

**Key features:**
- Declarative pipeline development
- Batch and streaming support

- Automated scaling and error handling
- Simplifies ETL and data engineering workflows



```python
%pip install dlt
import dlt
```

```python
rules = {
    "rule1" : "product_id is not null",
    "rule2" : "product_name is not null"
}
```

```python


@dlt.table()

@dlt.expect_all_or_drop(rules)
def dimproducts_stage():
  df = spark.readStream.table("databricks_catalog.silver.products_silver")
  return df   
```

```python
@dlt.view 

def dimproducts_view():
    df = spark.readStream.table("Live.dimproducts_stage")
    return df
```

```python
dlt.create_streaming_table("dim_products")
```

```python
dlt.apply_changes(
    target="dim_products",
    source="Live.dimproducts_stage",
    keys=["product_id"],
    sequence_by="product_id",
    stored_as_scd_type= 2
)
```
