# Load and Clean Regions Data from Bronze Layer

This section reads the `regions` table from the bronze layer in Azure Databricks, removes the `_rescued_data` column, and writes the cleaned data to the silver layer in Azure Data Lake using Delta format.


```python
df = spark.read.table("databricks_catalog.bronze.regions")

df.display()
```

```python
df = df.drop("_rescued_data")
```

```python
df.write.format("delta").mode("overwrite").save("abfss://silver@azuresadatalake.dfs.core.windows.net/regions")
```

```python
df = spark.read.format("delta").load("abfss://silver@azuresadatalake.dfs.core.windows.net/products")
df.display()
```

```python
%sql
create table databricks_catalog.silver.regions_silver using delta 
location "abfss://silver@azuresadatalake.dfs.core.windows.net/regions"
```
